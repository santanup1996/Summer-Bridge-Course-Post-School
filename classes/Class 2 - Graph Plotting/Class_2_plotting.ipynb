# Import


```python
# numpy         => to process vectors/matrices efficiently
# matplotlib    => For Plotting data

import numpy as np
import matplotlib.pyplot as plt
```

# Data generation

* Create a few x-points or, x-values as a vector/set/array
* Calculate y-values as a similar vector/set/array taking those x-values as input

## x-data

```python
# Manual array creation
x_array = np.array([0, 1, 2, 3, 4, 7, 9.5])
print(x_array)
```

```python
# FIXED NUMBER OF POINTS WITHIN AN INTERVAL WITH EQUAL SPACING
# I want 1000 points between x=-5, to x=5
npoints = 5
x_low = -2
x_high = 2
x_array = np.linspace(x_low, x_high, npoints)
print(x_array)
```

```python
# FIXED SPACED POINTS WITHIN AN INTERVAL
# I want a few points between x = -5 and x=5  so that, the spacing between 2 points is = 0.5
Δx = 2
x_low = -5
x_high = 5
x_array = np.arange(x_low, x_high, Δx)
print(x_array)
```

## y-data

```python
# y = 2*x
def linear_function(x):
    return 2*x


# FIXED NUMBER OF POINTS WITHIN AN INTERVAL WITH EQUAL SPACING
# I want 1000 points between x=-5, to x=5
npoints = 500
x_low = -5
x_high = 5
x_array = np.linspace(x_low, x_high, npoints)

y_array = linear_function(x_array)

# print(y_array)
```

# Plotting

```python
plt.plot(x_array, y_array, lw=4, color='red', ls='--')

plt.xlabel("x", size=26)
plt.ylabel("y", size=26)

plt.xlim(-3, 3)
plt.ylim(-6, 6)
```

# A new Function Plot

```python
# Function
# y = x^3 - 5x^2 + 2*x + 4
def f1(x):
    return (x**3.0) - 5*(x**2.0) + 2*x + 4

# Creating x-data
npoints = 1000
x_low = -5
x_high = 5
x_array = np.linspace(x_low, x_high, npoints)

# Calculating y-values
y_array = f1(x_array)

# Plotting
plt.plot(x_array, y_array, lw=3, color='blue', ls='-')

plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid()
```

# Many plots in one graph

```python
# We are using our previous x-data

# y = e^x
def f1(x):
    return np.exp(x)

# y = x^2
def f2(x):
    return x**2.0

# y = sin(x)
def f3(x):
    return np.sin(x)

y_exp_array = f1(x_array)

y_quad_array = f2(x_array)

y_sin_array = f3(x_array)

plt.plot(x_array, y_exp_array, lw=3, color='black', label='Exponential')
plt.plot(x_array, y_quad_array, lw=3, color='red', label='Quadratic')
plt.plot(x_array, y_sin_array, lw=3, color='blue', label='Sinusoidal')

plt.xlabel("x")
plt.ylabel("y")

plt.xlim(-2,2)
plt.ylim(-2, 5)

plt.legend()

plt.grid()
```
